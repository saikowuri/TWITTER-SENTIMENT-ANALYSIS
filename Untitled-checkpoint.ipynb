```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from tqdm import tqdm
```

```python
data = pd.read_csv('Tweets.csv')
data.head()
```

```python
data.shape
```

```python
data.dropna(subset=['tweet_location'], inplace=True)
data.shape
```

```python
data['tweet_location'].head(20)
```

```python
data.columns
```

```python
data.to_csv('Tweets_location.csv', index=False)
```

```python
geolocator = Nominatim(user_agent="sentiment")
```

```python
test = geolocator.geocode("San Mateo, CA & Las Vegas, NV")
type(test)
```

```python
from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode,
                      return_value_on_exception='North Pole',
                      max_retries=2,
                      min_delay_seconds=0.2,
                      swallow_exceptions=True)
latitudes = []
longitudes = []
for i in tqdm(range(len(data))):
    try:
        latitudes.append(geocode(data.tweet_location.iloc[i]).latitude)
        longitudes.append(geocode(data.tweet_location.iloc[i]).longitude)
    except:
        latitudes.append(0.0)
        longitudes.append(0.0)
        continue
data['latitude'] = latitudes
data['longitude'] = longitudes
data.head()
```
